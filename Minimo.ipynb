```python
import numpy as np

# Contador global para rastrear las evaluaciones de la distancia
distancia_calculada = 0

def calcular_distancia(p1, p2):
    """
    Calcula la distancia euclidiana entre dos puntos en 2D.
    Incrementa un contador global para rastrear el número de cálculos.
    """
    global distancia_calculada
    distancia_calculada += 1
    return np.sqrt((p2[0] - p1[0])**2 + (p2[1] - p1[1])**2)

def espacio_dividido(x_min, x_max, y_min, y_max, iteraciones, objetivo, max_dist_calculations):
    """
    Método de búsqueda que divide iterativamente el espacio para minimizar las evaluaciones de distancia.

    Parámetros:
        x_min, x_max: Límites del espacio en la dimensión x.
        y_min, y_max: Límites del espacio en la dimensión y.
        iteraciones: Número máximo de divisiones del espacio.
        objetivo: Coordenadas del punto objetivo (x, y).
        max_dist_calculations: Límite máximo de cálculos de distancia permitidos.

    Retorna:
        Mejor punto encontrado y su distancia.
    """
    global distancia_calculada
    distancia_calculada = 0  # Reiniciar el contador

    # Inicializar puntos de evaluación en los extremos y centro del espacio
    puntos = [
        ((x_min + x_max) / 2, (y_min + y_max) / 2),  # Centro
        (x_min, y_min),  # Esquina inferior izquierda
        (x_min, y_max),  # Esquina superior izquierda
        (x_max, y_min),  # Esquina inferior derecha
        (x_max, y_max),  # Esquina superior derecha
    ]

    # Evaluar las distancias iniciales
    distancias = [calcular_distancia(punto, objetivo) for punto in puntos]
    if distancia_calculada > max_dist_calculations:
        return None, None

    # Bucle de refinamiento
    for _ in range(iteraciones):
        # Seleccionar el punto más cercano como referencia
        idx_min = np.argmin(distancias)
        mejor_punto = puntos[idx_min]
        mejor_distancia = distancias[idx_min]

        # Reducir el espacio alrededor del mejor punto
        nueva_x_min = max(x_min, mejor_punto[0] - (x_max - x_min) / 4)
        nueva_x_max = min(x_max, mejor_punto[0] + (x_max - x_min) / 4)
        nueva_y_min = max(y_min, mejor_punto[1] - (y_max - y_min) / 4)
        nueva_y_max = min(y_max, mejor_punto[1] + (y_max - y_min) / 4)

        # Generar nuevos puntos en el subespacio
        puntos = [
            ((nueva_x_min + nueva_x_max) / 2, (nueva_y_min + nueva_y_max) / 2),  # Centro
            (nueva_x_min, nueva_y_min),
            (nueva_x_min, nueva_y_max),
            (nueva_x_max, nueva_y_min),
            (nueva_x_max, nueva_y_max),
        ]

        # Evaluar distancias en el subespacio
        distancias = [calcular_distancia(punto, objetivo) for punto in puntos]
        if distancia_calculada > max_dist_calculations:
            break

    # Devolver el mejor punto y la distancia encontrada
    idx_min = np.argmin(distancias)
    return puntos[idx_min], distancias[idx_min]

# Parámetros del problema
target_x, target_y = x1, y1  # Punto objetivo
x_min, x_max = -8, 8  # Límites en x
y_min, y_max = -8, 8  # Límites en y

# Ejecutar la búsqueda
mejor_punto, mejor_distancia = espacio_dividido(
    x_min, x_max, y_min, y_max, iteraciones=5, objetivo=(target_x, target_y), max_dist_calculations=20
)

# Imprimir resultados
print(f"Mejor punto: {mejor_punto}")
print(f"Distancia al objetivo: {mejor_distancia}")
print(f"Total de cálculos de distancia: {distancia_calculada}")
```
