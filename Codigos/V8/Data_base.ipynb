```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import StandardScaler
```

```python
# Cargar el archivo de resultados en un DataFrame de Pandas
df = pd.read_csv("resultados_filtrados.txt", sep="\t")
print(len(df))
# Mostrar las primeras filas del DataFrame
df.head()

New_df = df[["L","P","Inh","Trest","Trelative","C_h","Tipo_red","chi2"]]
New_df.head()
```

```python
L10=New_df[(New_df["L"]>=10) & (New_df["L"] <= 20)].sort_values(by="chi2", ascending=True).head(100)
L20=New_df[(New_df["L"]>=20) & (New_df["L"] <= 30)].sort_values(by="chi2", ascending=True).head(100)
L30=New_df[(New_df["L"]>=30) & (New_df["L"] <= 40)].sort_values(by="chi2", ascending=True).head(100)
L40=New_df[(New_df["L"]>=40) & (New_df["L"] <= 50)].sort_values(by="chi2", ascending=True).head(100)
L50=New_df[(New_df["L"]>=50) & (New_df["L"] <= 60)].sort_values(by="chi2", ascending=True).head(100)
L60=New_df[(New_df["L"]>=60) & (New_df["L"] <= 70)].sort_values(by="chi2", ascending=True).head(100)
L70=New_df[(New_df["L"]>=70) & (New_df["L"] <= 80)].sort_values(by="chi2", ascending=True).head(100)

T0=New_df[(New_df["Tipo_red"]==0)].sort_values(by="chi2", ascending=True).head(100)
T1=New_df[(New_df["Tipo_red"]==1)].sort_values(by="chi2", ascending=True).head(100)
T2=New_df[(New_df["Tipo_red"]==2)].sort_values(by="chi2", ascending=True).head(100)
T3=New_df[(New_df["Tipo_red"]==3)].sort_values(by="chi2", ascending=True).head(100)
T4=New_df[(New_df["Tipo_red"]==4)].sort_values(by="chi2", ascending=True).head(100)


correlation_matrix = New_df[["L","P","Inh","Trest","Trelative","C_h","chi2"]].corr()
plt.figure(figsize=(10, 8))
# Graficar la matriz de correlacin como un mapa de calor
sns.heatmap(correlation_matrix, annot=True, cmap="coolwarm", vmin=-1, vmax=1, square=True, cbar_kws={"shrink": .8})
plt.title("Matriz de Correlacin entre Parmetros")
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Specify the variables you want to plot
variables = ["L", "P", "Inh", "Trest", "Trelative", "C_h"]  # Add all relevant columns
chi2_values = T1["chi2"]

# Create the pairplot manually with color mapped to chi2
fig, axes = plt.subplots(len(variables), len(variables), figsize=(15, 15))
fig.subplots_adjust(hspace=0.05, wspace=0.05)

for i, var1 in enumerate(variables):
    for j, var2 in enumerate(variables):
        if i == j:
            # Diagonal: histograms for each variable
            axes[i, j].hist(T1[var1], bins=15, color='lightblue')
            axes[i, j].set_xlabel(var1)
            axes[i, j].set_ylabel(var2)
        else:
            # Off-diagonal: scatter plots with color mapped to chi2
            scatter = axes[i, j].scatter(T1[var2], T1[var1],s=1, c=chi2_values, cmap='viridis')
            if j == 0:
                axes[i, j].set_ylabel(var1)
            if i == len(variables) - 1:
                axes[i, j].set_xlabel(var2)

# Add colorbar for chi2 values
cbar = fig.colorbar(scatter, ax=axes, orientation='vertical', fraction=0.02, pad=0.02)
cbar.set_label('chi2')

plt.show()
```
